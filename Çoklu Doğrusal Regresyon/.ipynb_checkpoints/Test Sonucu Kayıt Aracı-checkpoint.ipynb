```python
import sys
import pandas as pd
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QGridLayout, QLabel, QLineEdit, QPushButton, QTableWidget, QTableWidgetItem, QFileDialog, QMessageBox, QComboBox


class VeriOnIslemeUygulamasi(QWidget):
    def __init__(self):
        super().__init__()
        self.veriler = []
        self.init_ui()

    def init_ui(self):
        layout = QVBoxLayout()
        grid_layout = QGridLayout()

        # Standartizasyon seçimi
        standartizasyon_label = QLabel("Standartizasyon:")
        grid_layout.addWidget(standartizasyon_label)

        self.standartizasyon_combo = QComboBox()
        self.standartizasyon_combo.addItem("Yapıldı")
        self.standartizasyon_combo.addItem("Yapılmadı")
        grid_layout.addWidget(self.standartizasyon_combo)

        # Diğer girdiler
        sayi_label = QLabel("Kaç kez yapıldı?")
        grid_layout.addWidget(sayi_label)

        self.sayi_input = QLineEdit()
        grid_layout.addWidget(self.sayi_input)

        anova_label = QLabel("ANOVA Testi:")
        grid_layout.addWidget(anova_label)

        self.anova_combo = QComboBox()
        self.anova_combo.addItem("Yapıldı")
        self.anova_combo.addItem("Yapılmadı")
        grid_layout.addWidget(self.anova_combo)

        cikarma_label = QLabel("Etkisiz Sütunlar:")
        grid_layout.addWidget(cikarma_label)

        self.cikarma_combo = QComboBox()
        self.cikarma_combo.addItem("Çıkarıldı")
        self.cikarma_combo.addItem("Çıkarılmadı")
        grid_layout.addWidget(self.cikarma_combo)

        etiket_label = QLabel("Etiketlendi mi?")
        grid_layout.addWidget(etiket_label)

        self.etiket_combo = QComboBox()
        self.etiket_combo.addItem("Etiketlendi")
        self.etiket_combo.addItem("Etiketlenmedi")
        grid_layout.addWidget(self.etiket_combo)

        korelasyon_label = QLabel("Korelasyon Testi:")
        grid_layout.addWidget(korelasyon_label)

        self.korelasyon_combo = QComboBox()
        self.korelasyon_combo.addItem("Yapıldı")
        self.korelasyon_combo.addItem("Yapılmadı")
        grid_layout.addWidget(self.korelasyon_combo)

        esik_label = QLabel("Eşik Değer:")
        grid_layout.addWidget(esik_label)

        self.esik_input = QLineEdit()
        grid_layout.addWidget(self.esik_input)

        r2_label = QLabel("R^^2 Skoru:")
        grid_layout.addWidget(r2_label)

        self.r2_input = QLineEdit()
        grid_layout.addWidget(self.r2_input)

        dogruluk_label = QLabel("Doğruluk Oranı (%):")
        grid_layout.addWidget(dogruluk_label)

        self.dogruluk_input = QLineEdit()
        grid_layout.addWidget(self.dogruluk_input)

        # Kaydet butonu
        kaydet_button = QPushButton("Kaydet")
        kaydet_button.clicked.connect(self.verileri_kaydet)
        grid_layout.addWidget(kaydet_button)

        # Excel Dosyasına Kaydet butonu
        excel_kaydet_button = QPushButton("Excel Dosyasına Kaydet")
        excel_kaydet_button.clicked.connect(self.excel_kaydet)
        grid_layout.addWidget(excel_kaydet_button)

        layout.addLayout(grid_layout)

        # Veri tablosu
        self.table_widget = QTableWidget()
        layout.addWidget(self.table_widget)

        self.setLayout(layout)

    def verileri_kaydet(self):
        veri = {
            'Standartizasyon': self.standartizasyon_combo.currentText(),
            'Kaç kez yapıldı?': self.sayi_input.text(),
            'ANOVA Testi': self.anova_combo.currentText(),
            'Etkisiz Sütunlar': self.cikarma_combo.currentText(),
            'Etiketlendi mi?': self.etiket_combo.currentText(),
            'Korelasyon Testi': self.korelasyon_combo.currentText(),
            'Eşik Değer': self.esik_input.text(),
            'R^^2 Skoru': self.r2_input.text(),
            'Doğruluk Oranı (%)': self.dogruluk_input.text(),
            'Verileri kaydet': '',
            'Excel Dosyasına Kaydet': ''
        }

        self.veriler.append(veri)
        self.listeyi_guncelle()

    def excel_kaydet(self):
        if len(self.veriler) == 0:
            QMessageBox.warning(self, "Hata", "Kaydedilecek veri bulunamadı.")
            return

        dosya_yolu, _ = QFileDialog.getSaveFileName(self, "Verileri Excel'e Kaydet", "", "Excel Dosyası (*.xlsx)")

        if dosya_yolu:
            df = pd.DataFrame(self.veriler)
            df = df.drop(['Verileri kaydet', 'Excel Dosyasına Kaydet'], axis=1)
            df.to_excel(dosya_yolu, index=False)

            QMessageBox.information(self, "Başarılı", "Veriler başarıyla kaydedildi.")

    def listeyi_guncelle(self):
        self.table_widget.clearContents()
        self.table_widget.setColumnCount(0)

        if len(self.veriler) == 0:
            self.table_widget.setRowCount(0)
            return

        column_count = len(self.veriler[0])
        self.table_widget.setColumnCount(column_count)
        self.table_widget.setHorizontalHeaderLabels(self.veriler[0].keys())

        row_count = len(self.veriler)
        self.table_widget.setRowCount(row_count)

        for row, veri in enumerate(self.veriler):
            for column, (key, value) in enumerate(veri.items()):
                item = QTableWidgetItem(value)
                self.table_widget.setItem(row, column, item)

        self.table_widget.resizeColumnsToContents()


if __name__ == '__main__':
    app = QApplication(sys.argv)
    uygulama = VeriOnIslemeUygulamasi()
    uygulama.show()
    sys.exit(app.exec_())
```
