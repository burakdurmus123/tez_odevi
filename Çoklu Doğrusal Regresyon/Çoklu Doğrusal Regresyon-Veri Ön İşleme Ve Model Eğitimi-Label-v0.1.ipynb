### 3- Veri kümemizde aykırı gözlem olup olmadığına bakıyoruz.

```python
from scipy import stats

modeller = ['Bayon', 'Kona', 'Taycan', 'Panamera']

aykirilar = {}

for _ in range(7):
    for model in modeller:
        model_fiyat = dataset[dataset['Model'] == model]['Fiyat']
        z_scores = stats.zscore(model_fiyat)
        outliers = model_fiyat[abs(z_scores) > 3]
        aykirilar[model] = outliers
        dataset = dataset[~dataset.index.isin(outliers.index)]

    y = pd.Series(y)
    for model, outliers in aykirilar.items():
        y = y[~y.index.isin(outliers.index)]

    dataset = dataset.dropna(how='all')

    dataset.reset_index(drop=True, inplace=True)
    y.reset_index(drop=True, inplace=True)

for model, outliers in aykirilar.items():
    print("Aykırı değerler -", model)
    print(outliers)
    print("Aykırı gözlem sayısı:", len(outliers))
    print()

print("*****Aykırı değerler*****")
print()
print(dataset.info())
print()
```

### 5- ANOVA ve Post Hoc Testi.

```python
import pandas as pd
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Sayısal sütunları seçin
sayısal_sütunlar = ["Model Yıl", "Araç Kilometresi"]
X_sayısal = dataset[sayısal_sütunlar].values

# Kategorik sütunları seçin
kategorik_sütunlar = list(set(dataset.columns) - set(sayısal_sütunlar) - {"Fiyat"})
X_kategorik = dataset[kategorik_sütunlar].values

# Bağımlı değişken (y) olarak "Fiyat" sütununu seçin
y = dataset["Fiyat"].values

# ANOVA testini uygulayın (Sayısal sütunlar)
f_istatistikleri_sayısal, p_değerleri_sayısal = [], []

for i in range(X_sayısal.shape[1]):
    gruplar = X_sayısal[:, i]
    f_istatistik, p_değer = f_oneway(*[y[gruplar == grup] for grup in set(gruplar)])
    f_istatistikleri_sayısal.append(f_istatistik)
    p_değerleri_sayısal.append(p_değer)

# ANOVA testini uygulayın (Kategorik sütunlar)
f_istatistikleri_kategorik, p_değerleri_kategorik = [], []

for i in range(X_kategorik.shape[1]):
    gruplar = X_kategorik[:, i]
    gruplar = gruplar.astype(str)  # Grupları string olarak dönüştürün
    f_istatistik, p_değer = f_oneway(*[y[gruplar == grup] for grup in set(gruplar)])
    f_istatistikleri_kategorik.append(f_istatistik)
    p_değerleri_kategorik.append(p_değer)

# Post hoc testini uygulayın (Tukey's HSD testi)
posthoc_sonuçları = {}

for sütun in kategorik_sütunlar:
    sütun_indeks = kategorik_sütunlar.index(sütun)
    tukey = pairwise_tukeyhsd(y, X_kategorik[:, sütun_indeks])
    posthoc_sonuçları[sütun] = tukey

# Sonuçları yazdırın
print("ANOVA Sonuçları (Sayısal Sütunlar):")
for i in range(len(f_istatistikleri_sayısal)):
    sütun_adı = sayısal_sütunlar[i]
    print("Sütun:", sütun_adı)
    print("F-istatistik değeri:", f_istatistikleri_sayısal[i])
    print("p-değeri:", p_değerleri_sayısal[i])
    print("-----------------------")

print("ANOVA Sonuçları (Kategorik Sütunlar):")
for i in range(len(f_istatistikleri_kategorik)):
    sütun_adı = kategorik_sütunlar[i]
    print("Sütun:", sütun_adı)
    print("F-istatistik değeri:", f_istatistikleri_kategorik[i])
    print("p-değeri:", p_değerleri_kategorik[i])
    print("-----------------------")

print("Post Hoc Testi (Tukey's HSD Testi) Sonuçları:")
for sütun, tukey in posthoc_sonuçları.items():
    print("Sütun:", sütun)
    print(tukey.summary())
    print("-----------------------")
    
  # Etkisiz değişkenleri depolamak için bir liste oluşturun
etkisiz_degiskenler = []

# Post hoc test sonuçlarını değerlendirin ve etkisiz değişkenleri belirleyin
for sütun, tukey in posthoc_sonuçları.items():
    gruplar = tukey.groupsunique
    p_değerleri = tukey.pvalues

    # Etkisiz değişkenleri belirleyin (p değeri 0.05'ten büyük olanlar)
    etkisiz_değişkenler_sütunu = [grup for grup, p in zip(gruplar, p_değerleri) if p > 0.05]
    
    if etkisiz_değişkenler_sütunu:
        etkisiz_degiskenler.append((sütun, etkisiz_değişkenler_sütunu))

# Etkisiz değişkenleri ekrana yazdırın
print("Etkisiz Değişkenler:")
for sütun, etkisiz_değişkenler_sütunu in etkisiz_degiskenler:
    print(f"{sütun}: {', '.join(etkisiz_değişkenler_sütunu)}")    
```

### 6- Korelasyon Testi

```python
from scipy.stats import pearsonr

# X'i pandas DataFrame'e dönüştürme
X = pd.DataFrame(X)

# Her bir sütunun "Fiyat" sütunu ile olan korelasyonunu test etme
korelasyonlar = []
for sutun in range(X.shape[1]):
    korelasyon, _ = pearsonr(X.iloc[:, sutun], y)
    korelasyonlar.append(korelasyon)

# Sütun adları ve korelasyon katsayılarını birleştirme
korelasyon_verileri = list(zip(X.columns, korelasyonlar))

# Korelasyonları mutlak değere göre büyükten küçüğe sıralama
korelasyon_verileri.sort(key=lambda x: abs(x[1]), reverse=True)

# Eşik değeri
esik_degeri = 0.5  # İstediğiniz eşik değerini burada belirleyin

# Eşik değerini geçen sütunları seçme
anlamli_sutunlar = [sutun for sutun, korelasyon in korelasyon_verileri if abs(korelasyon) > esik_degeri]

# Sadece anlamlı sütunları X değişkenine atama
X = X[anlamli_sutunlar]

# Sonuçları yazdırma
for sutun, korelasyon in korelasyon_verileri:
    print(f"{sutun}: Korelasyon= {korelasyon}")
```

# Veri Kümesini Eğitim Ve Test Seti Olarak Bölme

```python
from sklearn.model_selection import train_test_split

# Veri kümesini eğitim ve test verisi olarak bölmek
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=0)

# Eğitim verisini doğrulama/geliştirme verisi olarak bölmek
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.15, random_state =0)
```

# Modeli Eğitme İşlemi

```python
from sklearn.linear_model import LinearRegression
import joblib

regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Modelin katsayılarını ve kesme terimini alır.
coefficients = regressor.coef_
intercept = regressor.intercept_

# Modelin formülünü oluşturur.
formula = "y = {:.2f}".format(intercept)

# Katsayıları ve öznitelik isimlerini birleştirir.
for i, coefficient in enumerate(coefficients):
    coefficient_str = "{:.2f}".format(coefficient)  # Değişiklik burada
    formula += " + ({} * {})".format(coefficient_str, X_train.columns[i])

print("Modelin formülü:", formula)

# Modeli kaydet
joblib.dump(regressor, "Makine Öğrenim Modeli-Label v0.1.pkl")
```

# Eğitim Ve Test Sonuçları

```python
from scipy import stats
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Tahmin sonuçlarını elde etmek
y_pred_train = regressor.predict(X_train)
y_pred_val = regressor.predict(X_val)
y_pred_test = regressor.predict(X_test)

# R-Kare (R^2) Skoru
r2_train = r2_score(y_train, y_pred_train)
r2_val = r2_score(y_val, y_pred_val)
r2_test = r2_score(y_test, y_pred_test)

# Ortalama Mutlak Hata (MAE)
mae_train = mean_absolute_error(y_train, y_pred_train)
mae_val = mean_absolute_error(y_val, y_pred_val)
mae_test = mean_absolute_error(y_test, y_pred_test)

# Ortalama Kare Hata (MSE)
mse_train = mean_squared_error(y_train, y_pred_train)
mse_val = mean_squared_error(y_val, y_pred_val)
mse_test = mean_squared_error(y_test, y_pred_test)

# Kök Ortalama Kare Hata (RMSE)
rmse_train = np.sqrt(mse_train)
rmse_val = np.sqrt(mse_val)
rmse_test = np.sqrt(mse_test)

# P-değerleri
_, p_values_train = stats.ttest_ind(y_pred_train, y_train)
_, p_values_val = stats.ttest_ind(y_pred_val, y_val)
_, p_values_test = stats.ttest_ind(y_pred_test, y_test)

# Test tahminleri ve gerçek sonuçlar arasındaki farkın yüzdesel hesaplanması
percentage_diff_train = (y_pred_train - y_train) / y_train * 100
percentage_diff_val = (y_pred_val - y_val) / y_val * 100
percentage_diff_test = (y_pred_test - y_test) / y_test * 100

# Doğruluk oranlarının hesaplanması
accuracy_train = 100 - np.mean(np.abs(percentage_diff_train))
accuracy_val = 100 - np.mean(np.abs(percentage_diff_val))
accuracy_test = 100 - np.mean(np.abs(percentage_diff_test))

# Tahmin sonuçları ve gerçek değerlerin birleştirilerek görüntülenmesi
results_train = np.concatenate((y_pred_train.reshape(len(y_pred_train), 1), y_train.reshape(len(y_train), 1)), axis=1)
results_val = np.concatenate((y_pred_val.reshape(len(y_pred_val), 1), y_val.reshape(len(y_val), 1)), axis=1)
results_test = np.concatenate((y_pred_test.reshape(len(y_pred_test), 1), y_test.reshape(len(y_test), 1)), axis=1)

# Sonuçların yazdırılması
print("Eğitim Seti R^2 Skoru: {:.2f}".format(round(r2_train, 2)))
print("Eğitim Seti MAE: {:.2f}".format(round(mae_train, 2)))
print("Eğitim Seti MSE: {:.2f}".format(round(mse_train, 2)))
print("Eğitim Seti RMSE: {:.2f}".format(round(rmse_train, 2)))
print("Eğitim Seti P-değerleri: {:.2f}".format(round(p_values_train, 2)))
print("Eğitim Seti Doğruluk Oranı: {:.2f}%".format(accuracy_train))
print()
print("Eğitim Seti Tahmin Sonuçları ve Gerçek Değerler:")
print()
print(results_train[:10])
print()
print()
print("************************************************************")
print()

print("Doğrulama/Geliştirme Seti R^2 Skoru: {:.2f}".format(round(r2_val, 2)))
print("Doğrulama/Geliştirme Seti MAE: {:.2f}".format(round(mae_val, 2)))
print("Doğrulama/Geliştirme Seti MSE: {:.2f}".format(round(mse_val, 2)))
print("Doğrulama/Geliştirme Seti RMSE: {:.2f}".format(round(rmse_val, 2)))
print("Doğrulama/Geliştirme Seti P-değerleri: {:.2f}".format(round(p_values_val, 2)))
print("Doğrulama/Geliştirme Seti Doğruluk Oranı: {:.2f}%".format(accuracy_val))
print()
print("Doğrulama/Geliştirme Seti Tahmin Sonuçları ve Gerçek Değerler:")
print()
print(results_val[:10])
print()
print()
print("************************************************************")
print()

print("Test Seti R^2 Skoru: {:.2f}".format(round(r2_test, 2)))
print("Test Seti MAE: {:.2f}".format(round(mae_test, 2)))
print("Test Seti MSE: {:.2f}".format(round(mse_test, 2)))
print("Test Seti RMSE: {:.2f}".format(round(rmse_test, 2)))
print("Test Seti P-değerleri: {:.2f}".format(round(p_values_test, 2)))
print("Test Seti Doğruluk Oranı: {:.2f}%".format(accuracy_test))
print()
print("Test Seti Tahmin Sonuçları ve Gerçek Değerler:")
print()
print(results_test[:10])
```
